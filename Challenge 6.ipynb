```python
Task 3: Updating User Profiles (Delete File)
```

```python
import os
file_name = "temp.txt"

with open(file_name, "w") as file:
    file.write("This is a temporary file.")
print(f"{file_name} created successfully.")

if os.path.exists(file_name):
    print(f"{file_name} exists. Proceeding to delete...")

    os.remove(file_name)
    print(f"{file_name} deleted successfully.")

    if not os.path.exists(file_name):
        print(f"Confirmed: {file_name} no longer exists.")
    else:
        print(f"Error: {file_name} still exists.")
else:
    print(f"{file_name} does not exist. Nothing to delete.")
```

```python
Task 4: Log File Analysis (File Pointer)
```

```python
with open("activity.log", "w") as file:
    file.write("User A logged in\n")
    file.write("User B uploaded a file\n")
    file.write("User C logged out\n")
    file.write("User A updated profile\n")
    file.write("User D deleted a file\n")
print("Log entries written successfully.\n")

with open("activity.log", "r") as file:
    print("Full Log File:")
    print(file.read())

with open("activity.log", "r") as file:
    print("\n--- Using seek() ---")

    file.seek(0)
    print("\nFirst log entry:")
    print(file.readline().strip())

    file.seek(20)
    print("\nReading from position 20:")
    print(file.readline().strip())

    file.seek(50)
    print("\nReading from position 50:")
    print(file.readline().strip())
```

```python
Task 5: Folder Management Tool
```

```python
import os
folder_name = "temp_folder"

if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print("Folder created")
else:
    print("Folder already exists")

if os.path.exists(folder_name):
    os.rmdir(folder_name)
    print("Folder deleted successfully")
else:
    print("Folder does not exist")

if not os.path.exists(folder_name):
    print("Confirmed: Folder has been deleted")
else:
    print("Folder still exists")
```

```python
Task 6: File Compression Utility (ZIP Handling)
```

```python
import zipfile
import os

for i in range(1, 4):
    filename = f"file{i}.txt"
    with open(filename, "w") as f:
        f.write(f"This is content of {filename}")
    print(f"Created {filename}")

zip_filename = "archive.zip"

with zipfile.ZipFile(zip_filename, "w") as zipf:
    for i in range(1, 4):
        filename = f"file{i}.txt"
        zipf.write(filename)
        print(f"Added {filename} to {zip_filename}")

print(f"\nAll files compressed into {zip_filename}")

extract_folder = "extracted_files"

os.makedirs(extract_folder, exist_ok=True)

with zipfile.ZipFile(zip_filename, "r") as zipf:
    zipf.extractall(extract_folder)
    print(f"\nFiles extracted to '{extract_folder}' folder")
```
