# **Challenge 1: Practise Python Basics using String and Number Functions**

**Task 1: Reverse a string**

```python
def reverse_string(text):
    return text[::-1]

input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)

print("Reversed string:", reversed_string)
```

**Task 2: Count vowels in a string**

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

input_string = input("Enter a string: ")
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)
```

**Task 3: Check Palindrome Number**



```python
def is_palindrome_number(num):
    num_str = str(num)

    if num_str == num_str[::-1]:
        return "true"
    else:
        return "false"

number = int(input("Enter a number: "))
result = is_palindrome_number(number)
print(result)
```

#**Challenge 3: Python Data Handling with collections**

**Task 1: Manage Student Grade (Using Lists)**

```python
grades = [75, 85, 90, 88]
print("Current grades:", grades)

new_grade = int(input("Enter a new grade to add: "))
grades.append(new_grade)
print("Grades after adding:", grades)

highest = max(grades)
lowest = min(grades)
average = sum(grades) / len(grades)
print("Highest grade:", highest)
print("Lowest grade:", lowest)
print("Average grade:", average)

position = int(input("Enter the position to modify (starting from 0): "))
new_value = int(input("Enter the new grade: "))
grades[position] = new_value
print("Grades after modification:", grades)

remove_grade = int(input("Enter the grade to remove: "))
grades.remove(remove_grade)
print("Grades after removing:", grades)
```

**Task 3: Unique Voters (Using Sets)**

```python
voters1 = {"V101", "V102", "V103", "V104"}

voters1.add("V105")
voters1.remove("V102")

if "V103" in voters1:
    print("Voter already exists")
else:
    print("Voter not found")

voters2 = {"V103", "V106", "V107"}

all_voters = voters1 | voters2
print("Union:", all_voters)

common_voters = voters1 & voters2
print("Intersection:", common_voters)
```

# **Challenge 4: Python Logic Mastery with Control Flow**

**Task 1: Starry Triangle (Nested for loop)**

```python
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

**Task 2: Factorial Finder (While Loop)**

```python
x = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= x:
    factorial = factorial * i
    i += 1

print("The factorial is:", factorial)
```

**Task 3: Number Skipper (Continue Statement)**

```python
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    if i % 3 == 0:
        continue
    print(i)
```

**Task 4: Even sum calculator (Pass Statement)**

```python
number = int(input("Enter a number: "))
sum = 0

for i in range(1, number + 1):
    if i % 2 != 0:
        pass
    else:
        sum += i

print(sum)
```

#**Challenge 5: Python Function Mastery with Parameters & Arguments**

**Task 1: Area Calculation with Default Parameters**

```python
def rectangle_area(length, width):
    return length * width

def square_area(side=5):
    return side * side

rect = rectangle_area(10, 5)
sq1 = square_area()
sq2 = square_area(4)

print("Rectangle Area =", rect)
print("Square Area (default side) =", sq1)
print("Square Area (side = 4) =", sq2)
```

**Task 2: Lambda Expression Calculator**

```python
add = lambda a, b: a + b
sub = lambda a, b: a - b
mul = lambda a, b: a * b
div = lambda a, b: a / b

print("Addition:", add(2, 3))
print("Subtraction:", sub(5, 2))
print("Multiplication:", mul(4, 5))
print("Division:", div(10, 2))

numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x * x, numbers))

print("Squared values:", squared)
```

**Task 3: Person Details Printer**

```python
def person_details(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)

name = input("Enter your name: ")
age = input("Enter your age: ")
choice = input("Do you want to enter your city? (yes/no): ")
if choice == "yes":
    city = input("Enter your city: ")
else:
    city = "Unknown"

person_details(name, age, city)
```

**Task 4: Car Features (Arbitrary Keyword Arguments)**

```python
def display_car_features(**features):
    print("\nCar Features:")
    print("-" * 30)
    for key, value in features.items():
        print(f"{key}: {value}")

n = int(input("How many car features do you want to enter? "))
car_features = {}

for i in range(n):
    feature_name = input(f"Enter name of feature {i+1}: ")
    feature_value = input(f"Enter value of {feature_name}: ")
    car_features[feature_name] = feature_value

display_car_features(**car_features)
```
