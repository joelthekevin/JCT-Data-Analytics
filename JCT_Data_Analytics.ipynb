# **Challenge 1: Practise Python Basics using String and Number Functions**

**Task 1: Reverse a string**

```python
def reverse_string(text):
    return text[::-1]

input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)

print("Reversed string:", reversed_string)
```

**Task2: Count vowels in a string**

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

input_string = input("Enter a string: ")
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)
```

**Task 3: Check Palindrome Number**



```python
def is_palindrome_number(num):
    num_str = str(num)

    if num_str == num_str[::-1]:
        return "true"
    else:
        return "false"

number = int(input("Enter a number: "))
result = is_palindrome_number(number)
print(result)
```

# **Challenge 2: Python Basics using String and Number Functions**

**Task 1: Sentence String Operations**

```python
sentence = input("Enter a sentence: ")

uppercase_sentence = sentence.upper()
print("Uppercase Sentence:", uppercase_sentence)

replaced_sentence = sentence.replace(" ", "-")
print("Sentence with Hyphens:", replaced_sentence)

length = len(sentence)
print("Length of the Sentence:", length)

first_five = sentence[:5]
print("First 5 Characters:", first_five)
```

**Task 2: Temperature Converter**

```python
temperature = float(input("Enter the temperature in Celsius: "))
fahrenheit = (temperature * 9/5) + 32
print("The temperature in Fahrenheit is", fahrenheit)
```

**Task 3: String Slicing and formatting**

```python
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")

print("First 3 letters of first name:", first_name[0:3])
print("Last 3 letters of last name:", last_name[-3:])
print("Full name reversed is: "f"{last_name}, {first_name}")
```
