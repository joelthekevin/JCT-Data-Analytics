# **Challenge 1: Practise Python Basics using String and Number Functions**

**Task 1: Reverse a string**

```python
def reverse_string(text):
    return text[::-1]

input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)

print("Reversed string:", reversed_string)
```

**Task 2: Count vowels in a string**

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

input_string = input("Enter a string: ")
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)
```

**Task 3: Check Palindrome Number**



```python
def is_palindrome_number(num):
    num_str = str(num)

    if num_str == num_str[::-1]:
        return "true"
    else:
        return "false"

number = int(input("Enter a number: "))
result = is_palindrome_number(number)
print(result)
```

**Task 1: Sentence String Operations**

```python
sentence = input("Enter a sentence: ")

uppercase_sentence = sentence.upper()
print("Uppercase Sentence:", uppercase_sentence)

replaced_sentence = sentence.replace(" ", "-")
print("Sentence with Hyphens:", replaced_sentence)

length = len(sentence)
print("Length of the Sentence:", length)

first_five = sentence[:5]
print("First 5 Characters:", first_five)
```

#**Challenge 3: Python Data Handling with collections**

**Task 1: Manage Student Grade (Using Lists)**

```python
grades = [75, 85, 90, 88]
print("Current grades:", grades)

new_grade = int(input("Enter a new grade to add: "))
grades.append(new_grade)
print("Grades after adding:", grades)

highest = max(grades)
lowest = min(grades)
average = sum(grades) / len(grades)
print("Highest grade:", highest)
print("Lowest grade:", lowest)
print("Average grade:", average)

position = int(input("Enter the position to modify (starting from 0): "))
new_value = int(input("Enter the new grade: "))
grades[position] = new_value
print("Grades after modification:", grades)

remove_grade = int(input("Enter the grade to remove: "))
grades.remove(remove_grade)
print("Grades after removing:", grades)
```

**Task 2: Employee Data (Using Tuples)**

```python
employee = (101, "Jake", "Developer", 50000)

emp_id, name, department, salary = employee

print("Employee ID:", emp_id)
print("Name:", name)
print("Department:", department)
print("Salary:", salary)

print("\nName and Salary:")
print(name, salary)

print("\nLast two elements of the tuple:")
print(employee[2:4])
```

**Task 3: Unique Voters (Using Sets)**

```python
voters1 = {"V101", "V102", "V103", "V104"}

voters1.add("V105")
voters1.remove("V102")

if "V103" in voters1:
    print("Voter already exists")
else:
    print("Voter not found")

voters2 = {"V103", "V106", "V107"}

all_voters = voters1 | voters2
print("Union:", all_voters)

common_voters = voters1 & voters2
print("Intersection:", common_voters)
```
